```python
import requests
import matplotlib.pyplot as plt
```

```python
# Сформируем URL для доступа к сервису
url = 'http://127.0.0.1:5000/num/10'
# Вызовем сервис
r = requests.get(url)
```

```python
data = r.json()
data, type(data)
```

```python
# Функции для выполнения запроса к сервису

def make_url(cnt):
    base_url = 'http://127.0.0.1:5000/num/'
    res = base_url + str(cnt)
    return res

def get_data(cnt):
    url = make_url(cnt)
    r = requests.get(url)
    return r.json()
```

```python
cnt_list = [5, 10, 15]
for cnt in cnt_list:
    print('{} первых чисел последовательности Фибоначчи: {}'.format(cnt, get_data(cnt)))
```

```python
# Данные для графика
y_10 = get_data(10)
x_10 = list(range(1, len(y_10)+1))
```

```python
# Ось абсцисс - порядковые номера чисел
x_10
```

```python
# Ось ординат - числа
y_10
```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(x_10, y_10)
plt.xlabel('Ось абсцисс')
plt.ylabel('Ось ординат')
plt.title('Первые {} чисел последовательности'.format(len(y_10)))
plt.show()
```
